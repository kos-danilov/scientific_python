```python
"""
Dataset contains min and max weather temperatures for certain days. 
See desirable output in figure.
"""
import datetime
import pandas as pd
# Tips:
# use datetime.datetime
# use pivot_table in the end
```

```python
df = pd.read_csv("weather-raw.csv")
df
```

```python
formatted_df = pd.melt(
    frame = df,
    id_vars = ["id", "year", "month", "element"])
formatted_df
```

```python
formatted_df = formatted_df.dropna()
formatted_df
```

```python
formatted_df["day"] = formatted_df["variable"].str.extract("(\d)")
formatted_df
```

```python
formatted_df = formatted_df.drop(["variable"], axis = 1)
formatted_df
```

```python
formatted_df["date"] = ""
for i in range(len(formatted_df)):
    formatted_df.iloc[i, 6] = datetime.datetime(year = int(formatted_df.iloc[i, 1]), month = int(formatted_df.iloc[i, 2]), day = int(formatted_df.iloc[i, 5]))
formatted_df
```

```python
formatted_df = formatted_df.drop(["year", "month", "day"], axis = 1)
formatted_df
```

```python
formatted_df.pivot_table(index=['id', 'date'],
                            columns='element',
                            values='value',)
```
